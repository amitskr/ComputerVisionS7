```python
"""**Import Libraries**"""
import cv2 as cv
import skimage.io as io
```

```python
"""**Load Images**"""
image1 = cv.imread("/content/gdrive/MyDrive/content/gdrive/one-piece.jpg",
cv.IMREAD_GRAYSCALE)
image2 = cv.imread("/content/gdrive/MyDrive/content/gdrive/luffy.jpg",
cv.IMREAD_GRAYSCALE)
```

```python
"""**Innitialize keypoints detector and descriptor**"""
obr = cv.ORB_create()
cv.SIFT_create()
```

```python
"""**Detect keypoints detector and descriptors**"""
keyPoint1, desc1 = obr.detectAndCompute(image1,None)
keyPoint2, desc2 = obr.detectAndCompute(image2,None)
```

```python
"""**Innitialize Brute Force Matcher**"""
bf=cv.BFMatcher_create()
```

```python
"""**Match the Descriptor using Brute Force**"""
matches = bf.match(desc1,desc2)
```

```python
"""**Sort the Matches by Distance**"""
matches = sorted(matches,key=lambda x: x.distance)
```

```python
"""**Draw top 'n' matches**"""
n=20
result = cv.drawMatches(image1,keyPoint1,image2,keyPoint2,matches[:n],None,flags=2)
```

```python
"""**Display Result**"""
io.imshow(result)
```
